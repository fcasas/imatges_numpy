## Exercici 4

```python
# Mostreu-me amb exemples de diferents matrius, la regla fonamental de Broadcasting que diu :
# "les matrius es poden transmetre / broadcast si les seves dimensions coincideixen o si una de 
# les matrius té una mida d'1".


arr1 = np.random.randint(0,100,size = (5,5))
arr2 = np.random.randint(0,100,size = (2,5))
arr3 = np.random.randint(0,100,size = (5))
arr4 = np.random.randint(0,100,size = (4))
```

```python
# no podem operar amb matrius de mides diferents
arr1 * arr2
```

```python
# podem operar ja que les matrius comparteixen la mida de la primera dimensió
arr1 * arr3
```

```python
# podem operar ja que les matrius comparteixen la mida d una dimensió
arr2 * arr3
```

```python
# no podem operar amb matrius de mides diferents
arr2 * arr4
```
